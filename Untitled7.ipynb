<h4>This code is used for a ZYNQ fpga to connect the zynq processor to the fpga to send a webcam images to be encrypted using the fpga and to be returned back to the processor . The data transmission is done using AXI4 stream bus which interfaces the programmable logic through FIFOS as illustrated in the schematic   . AXI lite is used to send initializations and settings to the encryption core. </br>  Zynq processing unit has GP0 which is a general purpose AXI master port so that it can configure the DMA using AXI lite to setup the transfers and trigger them. The other interface is HP0 port is a high performance AXI slave port that is needed to access the DDR controller to allow the DMA to read and write from the DDR. DMA is used to pass the data directly between the DRAM and the programmable logic through the input and output FIFO using AXI4 Stream where the cipher core makes synchronization with the two FIFO modules using the three signals (valid, ready, tlast) where tlast denotes the the end of a plaintext and is used in the output side to free the bus and make it ready for a new reception . . MM2S represents the data direction from a memory maped device to the streaming side which is the cipher core</h4/>'))
